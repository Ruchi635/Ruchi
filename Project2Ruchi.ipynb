```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import os
from glob import glob

def create_spectrogram(filename, name, store_path):
    try:
        plt.interactive(False)
        print(f"Processing file: {filename}")
        clip, sample_rate = librosa.load(filename, sr=None)
        fig = plt.figure(figsize=[0.72, 0.72])
        ax = fig.add_subplot(111)
        ax.axes.get_xaxis().set_visible(False)
        ax.axes.get_yaxis().set_visible(False)
        ax.set_frame_on(False)
        S = librosa.feature.melspectrogram(y=clip, sr=sample_rate)
        librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
        save_path = os.path.join(store_path, name + '.jpg')
        plt.savefig(save_path, dpi=400, bbox_inches='tight', pad_inches=0)
        plt.close(fig)
        print(f"Spectrogram saved: {save_path}")
    except Exception as e:
        print(f"Error creating spectrogram for {filename}: {e}")

def convert_audio_files_to_jpg(input_directory, output_directory):
    # Ensure the output directory exists
    os.makedirs(output_directory, exist_ok=True)

    # Iterate through each genre subdirectory
    for genre_dir in os.listdir(input_directory):
        genre_path = os.path.join(input_directory, genre_dir)

        # Skip if not a directory
        if not os.path.isdir(genre_path):
            continue

        # Create a corresponding subdirectory in the output directory
        genre_output_path = os.path.join(output_directory, genre_dir)
        os.makedirs(genre_output_path, exist_ok=True)

        # Get all .au audio files in the genre directory
        audio_files = glob(os.path.join(genre_path, "*.au"))

        # Check if any audio files were found
        if not audio_files:
            print(f"No audio files found in {genre_dir}.")
            continue

        # Process each audio file
        for audio_file in audio_files:
            # Extract the base name for the spectrogram image
            base_name = os.path.splitext(os.path.basename(audio_file))[0]

            # Create the spectrogram
            create_spectrogram(audio_file, base_name, genre_output_path)

# Usage example:
# Main directory containing all genre subdirectories
input_directory = '/content/genres'  # Path to the main genres directory

# Directory where spectrograms will be saved
output_directory = '/content/spectrograms/train'  # Path to the main output directory

# Convert all audio files in each genre subdirectory to spectrograms
convert_audio_files_to_jpg(input_directory, output_directory)
```

```python
import os
import random
import shutil

# Define source and target directories
source_dir = r'/content/spectrograms/train'  # Path to the directory containing the original images
target_dir = r'/content/spectrograms/val'    # Path to the directory where the split images will be stored
split_ratio = 0.8

def Train_Test_Split(source_dir, target_dir, split_ratio):
    # Define target directories for train and validation sets
    train_dir = os.path.join(target_dir, 'train')
    val_dir = os.path.join(target_dir, 'val')

    # Create target directories if they don't exist
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)

    # Get the list of class directories
    classes = [d for d in os.listdir(source_dir) if os.path.isdir(os.path.join(source_dir, d))]

    # Debugging output to check found classes
    print(f"Classes found: {classes}")

    for class_name in classes:
        # Create class directories in train and val folders
        train_class_dir = os.path.join(train_dir, class_name)
        val_class_dir = os.path.join(val_dir, class_name)
        os.makedirs(train_class_dir, exist_ok=True)
        os.makedirs(val_class_dir, exist_ok=True)

        # Get list of images in the class directory
        class_dir = os.path.join(source_dir, class_name)
        images = [f for f in os.listdir(class_dir) if os.path.isfile(os.path.join(class_dir, f))]

        # Debugging output for images in each class
        print(f"Found {len(images)} images for class {class_name}")

        if not images:
            print(f"No images found for class {class_name}. Skipping...")
            continue

        # Shuffle the images
        random.shuffle(images)

        # Compute the split point
        split_point = int(len(images) * split_ratio)

        # Split the images into training and validation sets
        train_images = images[:split_point]
        val_images = images[split_point:]

        # Debugging output to check split sizes
        print(f"Splitting {len(images)} images into {len(train_images)} train and {len(val_images)} validation images for class {class_name}.")

        # Copy the images to the respective directories
        for img in train_images:
            shutil.copy(os.path.join(class_dir, img), os.path.join(train_class_dir, img))

        for img in val_images:
            shutil.copy(os.path.join(class_dir, img), os.path.join(val_class_dir, img))

    print("Data split completed successfully!")

# Call the function
Train_Test_Split(source_dir, target_dir, split_ratio)
```
